```python
import matplotlib.pyplot as plt

# Assuming train is your DataFrame and "is_fraud" is a column in it
y_train = train["is_fraud"]

# Calculate value counts
z = y_train.value_counts()

# Create a pie chart
fig, ax = plt.subplots()
ax.pie(z, labels=z.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'black'])

# Draw a circle to make it a donut chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')  
plt.tight_layout()

# Show the plot
plt.title('Fraudulent vs Non-Fraudulent Transactions\n')

plt.show()
```

# Logistic Regression

```python
y_train = Train["is_fraud"]
col = ["is_fraud","gender_encoded"]
X_train = Train.drop(columns=col)
y_test = Test["is_fraud"]
col = ["is_fraud","Job_Type_Mean_Encoded"]
X_test = Test.drop(columns=col)
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
print(model.score(X_test, y_test)*100,"%")
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
print(classifier.score(X_test, y_test)*100,"%")
```

# Decision Tree Classifier
